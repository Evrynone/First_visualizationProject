```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('result_voice_6month.csv')
df.dropna(inplace = True)
val1 = df["Predicted_Intents"].hist()
plt.suptitle("Predicted intents.")
plt.show()
val = df["Actual_Intents"].hist()
plt.suptitle("Actual intents.")
plt.show()

```

```python
Unique = list(df['Predicted_Intents'].unique())
print(Unique)
Val = []
for name in Unique:
  Val.append(df['Predicted_Intents'].value_counts()[name])
print(Val)
```

```python
plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams["figure.autolayout"] = True
```

```python
fig, ax = plt.subplots()
# Val = len(df)
p1 = ax.bar(Unique, Val, color=['red', 'blue', 'green', 'cyan', 'yellow', 'violet'])

ax.set_xlabel('PREDICTED_INTENTS')
ax.set_ylabel('Number of Intents')
ax.set_title('SMS Bundle Query DATA INTENT PREDICTIONS')
plt.grid(True, color = 'gray', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.axisbelow'] = True
for rect1 in p1:
    height = rect1.get_height()
    heightPercent = round(float(height/sum(Val)) * 100, 2)
    plt.annotate( "{}%".format(heightPercent),(rect1.get_x() + rect1.get_width()/2, height+.05),ha="center",va="bottom",fontsize=15)
    # plt.annotate( "{}".format(height),(rect1.get_x() + rect1.get_width()/2, height+.05),ha="center",va="bottom",fontsize=25)

plt.savefig("BarChartPrepaidSMSBundle.png")
plt.show()
```
